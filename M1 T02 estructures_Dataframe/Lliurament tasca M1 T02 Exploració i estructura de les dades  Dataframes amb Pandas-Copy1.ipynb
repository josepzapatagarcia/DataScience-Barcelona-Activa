### Taula de les aerolínies amb més endarreriments acumulats


```python
#retrasos = df[['Origin', 'Dest', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Delayed', 'ArrDelay']].copy()
retrasos = df[['Origin', 'Dest',  'UniqueCarrier','ArrDelay']].copy()
retrasos_por_compañía = retrasos.groupby('UniqueCarrier').aggregate(sum)
retrasos_por_compañía.sort_values(by='ArrDelay', axis=0, ascending=False)
```

### Quins són els vols més llargs? I els més endarrerits?


```python
df['vuelos'] = df['Origin'] + ' - ' + df['Dest']
df['vuelos'] 
```

```python
#retrasos = df[['vuelos', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Distance', 'AirTime', 'Delayed', 'ArrDelay']].copy()
vols_mes_llargs = df[['vuelos', 'UniqueCarrier', 'Distance', 'AirTime',]].copy()
vols_mes_llargs_por_compañía = vols_mes_llargs.groupby('vuelos').mean()
vols_mes_llargs_por_compañía.sort_values(by='AirTime', axis=0, ascending=False)

```

```python
vols_mes_llargs_por_compañía = vols_mes_llargs.groupby('vuelos').describe()
vols_mes_llargs_por_compañía.head()
```

### I els més endarrerits?

```python
#retrasos = df[['vuelos', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Distance', 'AirTime', 'Delayed', 'ArrDelay']].copy()
vols_mes_retrasats = df[['vuelos', 'UniqueCarrier',  'ArrDelay']].copy()
vols_mes_retrasats_por_compañía = vols_mes_retrasats.groupby('vuelos').mean()
vols_mes_retrasats_por_compañía.sort_values(by='ArrDelay', axis=0, ascending=False)
```

```python
vols_mes_retrasats_por_compañía = vols_mes_retrasats.groupby('vuelos').describe()
vols_mes_retrasats_por_compañía.head()
```
